Реализация алгоритма на Python

```python
import operator
from random import *
ln = 1500
m = [randint(1, 1499) for _ in range(ln)]
def merge(left, right, compare):
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if compare(left[i], right[j]):
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    while i < len(left):
        result.append(left[i])
        i += 1
    while j < len(right):
        result.append(right[j])
        j += 1
    return result

def merge_sort(L, compare=operator.lt):
    if len(L) < 2:
        return L[:]
    else:
        middle = int(len(L) / 2)
        left = merge_sort(L[:middle], compare)
        right = merge_sort(L[middle:], compare)
        return merge(left, right, compare)

merge_sort(m)
```

Реализация алгоритма на Cython

```python
%%cython
import copy
import operator
from random import *
ln = 1500
m = [randint(1, 1499) for _ in range(ln)]

def merge_cython(left, right, compare):
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if compare(left[i], right[j]):
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    while i < len(left):
        result.append(left[i])
        i += 1
    while j < len(right):
        result.append(right[j])
        j += 1
    return result

def merge_sort_cython(L, compare=operator.lt):
    if len(L) < 2:
        return L[:]
    else:
        middle = int(len(L) / 2)
        left = merge_sort_cython(L[:middle], compare)
        right = merge_sort_cython(L[middle:], compare)
        return merge_cython(left, right, compare)

merge_sort_cython(m)
```
